# Fundamentos de DataFrames en Pandas

## Introducción

En este notebook exploraremos los conceptos fundamentales de **DataFrames**, una de las estructuras de datos más importantes en Python para el análisis de datos y la inteligencia artificial.

### ¿Qué es un DataFrame?

Un DataFrame es una estructura de datos bidimensional (como una tabla o una hoja de cálculo) que:
- Organiza datos en filas y columnas
- Permite almacenar diferentes tipos de datos en cada columna
- Facilita la manipulación y análisis de datos
- Es la base para muchas operaciones de Machine Learning

### ¿Por qué es importante en IA?

Los DataFrames son esenciales porque:
- Permiten cargar y explorar datasets
- Facilitan la limpieza y preparación de datos
- Son compatibles con bibliotecas de ML como scikit-learn
- Hacen el análisis exploratorio de datos más eficiente

### **Creación de DataFrames**

```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np

# Verificar la versión de pandas
print(f"Versión de pandas: {pd.__version__}")
print(f"Versión de numpy: {np.__version__}")
```

```python
# Crear un DataFrame desde un diccionario
# Cada clave del diccionario se convierte en una columna

datos_estudiantes = {
    'nombre': ['Ana', 'Carlos', 'María', 'Juan', 'Laura'],
    'edad': [22, 24, 23, 25, 22],
    'carrera': ['Ingeniería', 'Matemáticas', 'Ingeniería', 'Física', 'Matemáticas'],
    'promedio': [8.5, 9.2, 8.8, 7.9, 9.5]
}

df_estudiantes = pd.DataFrame(datos_estudiantes)

# Mostrar el DataFrame
print("DataFrame de Estudiantes:")
print(df_estudiantes)
```

```python
# Crear un DataFrame desde listas
# Podemos pasar una lista de listas y especificar los nombres de las columnas

datos_lista = [
    ['Pedro', 21, 'Biología', 8.3],
    ['Sofia', 23, 'Química', 9.0],
    ['Diego', 22, 'Biología', 8.7]
]

# Especificamos los nombres de las columnas
columnas = ['nombre', 'edad', 'carrera', 'promedio']

df_desde_lista = pd.DataFrame(datos_lista, columns=columnas)

print("DataFrame creado desde listas:")
print(df_desde_lista)
```

### **Exploración básica**

```python
# Explorar las primeras y últimas filas del DataFrame

print("Primeras 3 filas del DataFrame:")
print(df_estudiantes.head(3))
print("\n" + "="*50 + "\n")

print("Últimas 2 filas del DataFrame:")
print(df_estudiantes.tail(2))
print("\n" + "="*50 + "\n")

print("Dimensiones del DataFrame (filas, columnas):")
print(f"Shape: {df_estudiantes.shape}")
print(f"Número de filas: {df_estudiantes.shape[0]}")
print(f"Número de columnas: {df_estudiantes.shape[1]}")
```

```python
# Obtener información general del DataFrame

print("Información general del DataFrame:")
print(df_estudiantes.info())
print("\n" + "="*50 + "\n")

print("Estadísticas descriptivas de las columnas numéricas:")
print(df_estudiantes.describe())
print("\n" + "="*50 + "\n")

print("Tipos de datos de cada columna:")
print(df_estudiantes.dtypes)
```

### **Selección y filtrado**

```python
# Seleccionar columnas individuales o múltiples

print("Seleccionar una sola columna (retorna una Serie):")
print(df_estudiantes['nombre'])
print(f"\nTipo: {type(df_estudiantes['nombre'])}")
print("\n" + "="*50 + "\n")

print("Seleccionar múltiples columnas (retorna un DataFrame):")
columnas_seleccionadas = df_estudiantes[['nombre', 'promedio']]
print(columnas_seleccionadas)
print(f"\nTipo: {type(columnas_seleccionadas)}")
print("\n" + "="*50 + "\n")

print("Acceder a una columna usando notación de punto:")
print(df_estudiantes.carrera)
```

```python
# Seleccionar filas usando iloc (por índice numérico) y loc (por etiqueta)

print("Seleccionar la primera fila usando iloc (índice numérico):")
print(df_estudiantes.iloc[0])
print("\n" + "="*50 + "\n")

print("Seleccionar las filas 1 a 3 usando iloc:")
print(df_estudiantes.iloc[1:4])
print("\n" + "="*50 + "\n")

print("Seleccionar fila y columna específica usando iloc:")
print(f"Nombre del estudiante en fila 2: {df_estudiantes.iloc[2, 0]}")
print("\n" + "="*50 + "\n")

print("Seleccionar usando loc (por etiqueta de índice):")
print(df_estudiantes.loc[0:2, ['nombre', 'edad']])
```

```python
# Filtrar filas basándose en condiciones

print("Estudiantes con promedio mayor a 9.0:")
estudiantes_sobresalientes = df_estudiantes[df_estudiantes['promedio'] > 9.0]
print(estudiantes_sobresalientes)
print("\n" + "="*50 + "\n")

print("Estudiantes de Ingeniería:")
estudiantes_ingenieria = df_estudiantes[df_estudiantes['carrera'] == 'Ingeniería']
print(estudiantes_ingenieria)
print("\n" + "="*50 + "\n")

print("Estudiantes con edad 22 Y promedio mayor a 8.5:")
filtro_multiple = df_estudiantes[(df_estudiantes['edad'] == 22) & (df_estudiantes['promedio'] > 8.5)]
print(filtro_multiple)
print("\n" + "="*50 + "\n")

print("Estudiantes de Matemáticas O Física:")
filtro_or = df_estudiantes[(df_estudiantes['carrera'] == 'Matemáticas') | (df_estudiantes['carrera'] == 'Física')]
print(filtro_or)
```

### **Manipulación de Datos**

```python
# Agregar nuevas columnas al DataFrame

print("DataFrame original:")
print(df_estudiantes)
print("\n" + "="*50 + "\n")

# Agregar una nueva columna calculada
df_estudiantes['beca'] = df_estudiantes['promedio'] >= 9.0
print("DataFrame con columna 'beca' (promedio >= 9.0):")
print(df_estudiantes)
print("\n" + "="*50 + "\n")

# Agregar otra columna con valores constantes
df_estudiantes['semestre'] = 5
print("DataFrame con columna 'semestre':")
print(df_estudiantes)
print("\n" + "="*50 + "\n")

# Eliminar una columna
df_estudiantes_sin_semestre = df_estudiantes.drop('semestre', axis=1)
print("DataFrame después de eliminar la columna 'semestre':")
print(df_estudiantes_sin_semestre)
print("\n" + "="*50 + "\n")

# Eliminar columnas permanentemente (usando inplace=True)
df_estudiantes.drop('semestre', axis=1, inplace=True)
print("DataFrame después de eliminar 'semestre' permanentemente:")
print(df_estudiantes)
```

```python
# Ordenar el DataFrame por una o más columnas

print("Ordenar por promedio (ascendente):")
df_ordenado_asc = df_estudiantes.sort_values('promedio')
print(df_ordenado_asc)
print("\n" + "="*50 + "\n")

print("Ordenar por promedio (descendente):")
df_ordenado_desc = df_estudiantes.sort_values('promedio', ascending=False)
print(df_ordenado_desc)
print("\n" + "="*50 + "\n")

print("Ordenar por múltiples columnas (carrera y luego promedio):")
df_ordenado_multiple = df_estudiantes.sort_values(['carrera', 'promedio'], ascending=[True, False])
print(df_ordenado_multiple)
print("\n" + "="*50 + "\n")

print("Ordenar por índice:")
df_ordenado_index = df_ordenado_desc.sort_index()
print(df_ordenado_index)
```

```python
# Crear un DataFrame con valores nulos para practicar

datos_con_nulos = {
    'nombre': ['Alberto', 'Beatriz', 'Carmen', 'Daniel', 'Elena'],
    'edad': [21, np.nan, 24, 23, np.nan],
    'carrera': ['Informática', 'Química', np.nan, 'Informática', 'Química'],
    'promedio': [8.5, 9.1, 8.0, np.nan, 9.3]
}

df_nulos = pd.DataFrame(datos_con_nulos)

print("DataFrame con valores nulos:")
print(df_nulos)
print("\n" + "="*50 + "\n")

print("Detectar valores nulos:")
print(df_nulos.isnull())
print("\n" + "="*50 + "\n")

print("Contar valores nulos por columna:")
print(df_nulos.isnull().sum())
print("\n" + "="*50 + "\n")

print("Eliminar filas con cualquier valor nulo:")
df_sin_nulos = df_nulos.dropna()
print(df_sin_nulos)
print("\n" + "="*50 + "\n")

print("Rellenar valores nulos con un valor específico:")
df_rellenado = df_nulos.fillna({'edad': 22, 'carrera': 'Sin especificar', 'promedio': 0})
print(df_rellenado)
```

### **Operaciones Estadísticas Básicas**

```python
# Realizar operaciones estadísticas sobre el DataFrame

print("Promedio de edad:")
print(df_estudiantes['edad'].mean())
print("\n" + "="*50 + "\n")

print("Promedio más alto:")
print(df_estudiantes['promedio'].max())
print("\n" + "="*50 + "\n")

print("Promedio más bajo:")
print(df_estudiantes['promedio'].min())
print("\n" + "="*50 + "\n")

print("Mediana de los promedios:")
print(df_estudiantes['promedio'].median())
print("\n" + "="*50 + "\n")

print("Desviación estándar de los promedios:")
print(df_estudiantes['promedio'].std())
print("\n" + "="*50 + "\n")

print("Suma total de edades:")
print(df_estudiantes['edad'].sum())
print("\n" + "="*50 + "\n")

print("Contar valores únicos en carrera:")
print(df_estudiantes['carrera'].value_counts())
```

```python
# Agrupar datos y realizar operaciones agregadas

print("Promedio de calificaciones por carrera:")
promedio_por_carrera = df_estudiantes.groupby('carrera')['promedio'].mean()
print(promedio_por_carrera)
print("\n" + "="*50 + "\n")

print("Múltiples estadísticas por carrera:")
stats_por_carrera = df_estudiantes.groupby('carrera')['promedio'].agg(['mean', 'min', 'max', 'count'])
print(stats_por_carrera)
print("\n" + "="*50 + "\n")

print("Agrupar por edad y calcular promedio:")
por_edad = df_estudiantes.groupby('edad')['promedio'].mean()
print(por_edad)
print("\n" + "="*50 + "\n")

print("Agrupar y obtener todas las columnas numéricas:")
resumen_por_carrera = df_estudiantes.groupby('carrera').agg({
    'edad': 'mean',
    'promedio': ['mean', 'max']
})
print(resumen_por_carrera)
```

```python
# Aplicar funciones a columnas o filas del DataFrame

# Función simple para categorizar promedios
def categorizar_promedio(promedio):
    if promedio >= 9.0:
        return 'Excelente'
    elif promedio >= 8.0:
        return 'Bueno'
    else:
        return 'Regular'

print("Aplicar función a una columna:")
df_estudiantes['categoria'] = df_estudiantes['promedio'].apply(categorizar_promedio)
print(df_estudiantes)
print("\n" + "="*50 + "\n")

# Usar funciones lambda
print("Aplicar función lambda para convertir edad a meses:")
df_estudiantes['edad_meses'] = df_estudiantes['edad'].apply(lambda x: x * 12)
print(df_estudiantes[['nombre', 'edad', 'edad_meses']])
print("\n" + "="*50 + "\n")

# Aplicar función a múltiples columnas
print("Aplicar función a todo el DataFrame (operación numérica):")
df_numerico = df_estudiantes[['edad', 'promedio']].apply(lambda x: x * 2)
print(df_numerico)
print("\n" + "="*50 + "\n")

# Limpiar columnas temporales
df_estudiantes.drop(['categoria', 'edad_meses'], axis=1, inplace=True)
```

### **Visualización Simple con Pandas**

```python
# Pandas incluye capacidades básicas de visualización usando matplotlib

import matplotlib.pyplot as plt

# Configurar el estilo de las gráficas
plt.style.use('default')

# Gráfico de barras: Promedio por carrera
print("Gráfico de barras: Promedio por carrera")
df_estudiantes.groupby('carrera')['promedio'].mean().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Promedio de Calificaciones por Carrera')
plt.xlabel('Carrera')
plt.ylabel('Promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Histograma: Distribución de promedios
print("\nHistograma: Distribución de promedios")
df_estudiantes['promedio'].plot(kind='hist', bins=5, color='lightgreen', edgecolor='black')
plt.title('Distribución de Promedios')
plt.xlabel('Promedio')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# Gráfico de dispersión: Edad vs Promedio
print("\nGráfico de dispersión: Edad vs Promedio")
df_estudiantes.plot(kind='scatter', x='edad', y='promedio', color='coral', s=100)
plt.title('Relación entre Edad y Promedio')
plt.xlabel('Edad')
plt.ylabel('Promedio')
plt.tight_layout()
plt.show()
```

## Ejercicio Práctico: Análisis de Datos de Ventas

Ahora vamos a poner en práctica todo lo aprendido con un ejercicio que integra múltiples conceptos.

**Escenario:**
Eres analista de datos en una tienda y debes analizar las ventas del último mes.

**Tareas a realizar:**
1. Crear un DataFrame con datos de ventas
2. Explorar los datos básicos
3. Filtrar ventas mayores a $500
4. Calcular el total de ventas por producto
5. Identificar el producto más vendido
6. Calcular estadísticas por categoría
7. Visualizar los resultados


```python
# PASO 1: Crear el DataFrame de ventas
datos_ventas = {
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Laptop', 'Mouse', 'Audífonos', 'Teclado', 'Monitor', 'Laptop'],
    'categoria': ['Computadoras', 'Accesorios', 'Accesorios', 'Computadoras', 'Computadoras', 'Accesorios', 'Accesorios', 'Accesorios', 'Computadoras', 'Computadoras'],
    'precio': [800, 25, 45, 300, 850, 20, 60, 50, 320, 780],
    'cantidad': [2, 5, 3, 1, 1, 8, 4, 2, 2, 3],
    'vendedor': ['Ana', 'Carlos', 'Ana', 'María', 'Carlos', 'Ana', 'María', 'Carlos', 'Ana', 'María']
}

df_ventas = pd.DataFrame(datos_ventas)

print("DataFrame de Ventas:")
print(df_ventas)
print("\n" + "="*50 + "\n")

# PASO 2: Explorar datos básicos
print("Información del DataFrame:")
print(df_ventas.info())
print("\n" + "="*50 + "\n")

print("Primeras filas:")
print(df_ventas.head())
print("\n" + "="*50 + "\n")

print("Estadísticas descriptivas:")
print(df_ventas.describe())
```

```python
# PASO 3: Agregar columna de total de venta
df_ventas['total_venta'] = df_ventas['precio'] * df_ventas['cantidad']

print("DataFrame con columna 'total_venta':")
print(df_ventas)
print("\n" + "="*50 + "\n")

# PASO 4: Filtrar ventas mayores a $500
ventas_grandes = df_ventas[df_ventas['total_venta'] > 500]
print("Ventas mayores a $500:")
print(ventas_grandes)
print("\n" + "="*50 + "\n")

# PASO 5: Calcular el total de ventas por producto
ventas_por_producto = df_ventas.groupby('producto')['total_venta'].sum().sort_values(ascending=False)
print("Total de ventas por producto (ordenado):")
print(ventas_por_producto)
print("\n" + "="*50 + "\n")

# PASO 6: Identificar el producto más vendido
producto_mas_vendido = ventas_por_producto.idxmax()
total_mas_vendido = ventas_por_producto.max()
print(f"Producto más vendido: {producto_mas_vendido}")
print(f"Total de ventas: ${total_mas_vendido}")
```

```python
# PASO 7: Calcular estadísticas por categoría
print("Estadísticas de ventas por categoría:")
stats_categoria = df_ventas.groupby('categoria').agg({
    'total_venta': ['sum', 'mean', 'count'],
    'cantidad': 'sum'
})
print(stats_categoria)
print("\n" + "="*50 + "\n")

# PASO 8: Análisis por vendedor
print("Total de ventas por vendedor:")
ventas_por_vendedor = df_ventas.groupby('vendedor')['total_venta'].sum().sort_values(ascending=False)
print(ventas_por_vendedor)
print("\n" + "="*50 + "\n")

# PASO 9: Visualización de resultados
import matplotlib.pyplot as plt

# Gráfico 1: Ventas por producto
plt.figure(figsize=(10, 5))
ventas_por_producto.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Total de Ventas por Producto', fontsize=14, fontweight='bold')
plt.xlabel('Producto')
plt.ylabel('Total de Ventas ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico 2: Ventas por vendedor
plt.figure(figsize=(8, 5))
ventas_por_vendedor.plot(kind='barh', color='coral', edgecolor='black')
plt.title('Total de Ventas por Vendedor', fontsize=14, fontweight='bold')
plt.xlabel('Total de Ventas ($)')
plt.ylabel('Vendedor')
plt.tight_layout()
plt.show()
```
